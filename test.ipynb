```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def gain_to_fluct_ratio_for_odds(g, q):
    return (g - 1) / np.sqrt((q * g) - (g**2))

# Define the function
def gain_to_fluct_ratio_for_probs(g, p):
    return (g - 1) / (g * np.sqrt((1/p) - 1))

# Generate x values
g= np.linspace(0, 4, 400)

# Values of 'a' to plot
q_values = [1.5, 2, 2.5, 3]

p_values = [0.1, 0.3, 0.5, 0.7, 0.9]

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the function for different values of 'a'
for q in q_values:
    y = gain_to_fluct_ratio_for_odds(g, q)
    plt.plot(g, y, label=f'q = {q}')

# Add labels and title
plt.xlabel('g')
plt.ylabel('f(g)')
plt.title('Plot of f(g) = (g-1) / sqrt(q*x - g^2) for different values of q')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def gain_to_fluct_ratio_for_probs(g, p):
    return (g - 1) / (g * np.sqrt((1/p) - 1))

# Generate x values
g= np.linspace(1, 4, 400)

p_values = [0.1, 0.3, 0.5, 0.7, 0.99]

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the function for different values of 'a'
for p in p_values:
    y = gain_to_fluct_ratio_for_probs(g, p)
    plt.plot(g, y, label=f'p = {p}')

# Add labels and title
plt.xlabel('g')
plt.ylabel('f(g)')
plt.title('Plot of f(g) = (g-1) / sqrt(q*x - g^2) for different values of q')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
